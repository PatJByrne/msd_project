```python
from math import pi as PI
from math import sin, cos
import random

import IPython.html
from IPython.html import widgets 
from IPython.display import display, clear_output

from plotly.widgets import GraphWidget
from plotly.graph_objs import *
import plotly.plotly as py
import plotly.tools as tls
```

```python
import Graph_dictionary
import distance_from_lat_long

class trip(object):
    def __init__(self,st = None, nxt = None,num =1):
        self.st = st
        self.nxt = nxt
        self.num = num
        
    def increment(self):
        self.num += 1
        
(PT,X,Y) = distance_from_lat_long.polar_to_cartesian()
team_graph = Graph_dictionary.team_grapher()
teams = team_graph.keys()

cpoints = Scatter(x=X, y=Y,mode = 'markers',name = 'Checkpoints',text = PT.astype(str),marker = Marker(size = 12,color ='rgba(0,255,0,1)'))
def trips(trip_num):
    out_lines = []
    in_lines = []
    departures = {}
    arrivals = {}

    for t in teams:
        if len(team_graph[t])<=trip_num:
            continue
        else:
            print trip_num,len(team_graph[t])
            prv = team_graph[t][trip_num].prv
            cur = team_graph[t][trip_num].num
            nxt = team_graph[t][trip_num].nxt
            arr_key = str(prv)+'-'+str(cur)
            dep_key = str(cur)+'-'+str(nxt)

            if arr_key not in arrivals.keys():
                arrivals[arr_key] =trip(prv,cur)
            else: arrivals[arr_key].increment()
            if dep_key not in departures.keys():
                departures[dep_key] =trip(cur,nxt)
            else: departures[dep_key].increment()
        
    for d in departures.keys():
        clr = 'rgba(255,0,0,%0.1f)' % 0.5**(1/departures[d].num)
        out_lines.append(Scatter(x=[X[departures[d].st-100], X[departures[d].nxt-100]],
                                   y=[Y[departures[d].st-100], Y[departures[d].nxt-100]], mode='lines',
                                   line=Line(color=clr, width=1+departures[d].num/4.)))

        
    for a in arrivals.keys():
        clr = 'rgba(0,0,255,%0.1f)' % 0.5**(1/arrivals[a].num)
        out_lines.append(Scatter(x=[X[arrivals[a].st-100], X[arrivals[a].nxt-100]],
                                    y=[Y[arrivals[a].st-100], Y[arrivals[a].nxt-100]], mode='lines',
                                    line=Line(color=clr, width=1+arrivals[a].num/4.)))


    data = Data(out_lines+in_lines+[cpoints])
    return(data)


step = widgets.IntSlider()
(step.min,step.max)= (0,38)

step.value = 12
step.description = 'Checkpoint #'

g = GraphWidget('https://plot.ly/~byrnsey24/72')

def plot_trips(_):
    data = trips(step.value)

    g.restyle({'x': [[]], 'y': [[]]})
    g.add_traces(data)
    
    layout = Layout(
        autosize=False,
        width=200,
        height=180)
    #fig = Figure(data=data, layout=layout)
    #plot_url = py.plot(fig, filename='Racoongaine_map')#,auto_open=False)   
step.on_trait_change(plot_trips, 'value')

display(step)
display(g)
```
